## 1. Doing the Changes for Pied Kingfisher Images

```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the 2 images
base_image = cv2.imread(r'W:\University of Moratuwa\Academics\Semester 5\EN3160 - Image Processing and Machine Vision\Assignments\Assignment_2\Bird_Base.JPG')
add_image = cv2.imread(r'W:\University of Moratuwa\Academics\Semester 5\EN3160 - Image Processing and Machine Vision\Assignments\Assignment_2\Bird_Add.JPG')

# Define the four points on the base image where you want to superimpose the other image
# [y , x]
# Top left, Top right, Bottom right, Bottom left
base_points = np.array([[2000, 750], [3500, 750], [3500, 1750], [2000, 1750]], dtype=np.float32)

# Define the four corners of the add image
add_points = np.array([[0, 0], [add_image.shape[1], 0], [add_image.shape[1], add_image.shape[0]], [0, add_image.shape[0]]], dtype=np.float32)

# Compute the homography matrix
homography_matrix, _ = cv2.findHomography(add_points, base_points)

# Warp the add image to fit the base image
add_warped = cv2.warpPerspective(add_image, homography_matrix, (base_image.shape[1], base_image.shape[0]))

# Create a mask for the add image
add_mask_inv = np.zeros_like(base_image)
cv2.fillPoly(add_mask_inv, [base_points.astype(int)], (255, 255, 255))

# Invert the mask
add_mask_inv_inv = cv2.bitwise_not(add_mask_inv)

# Blend the add image onto the base image
result = cv2.bitwise_and(base_image, add_mask_inv_inv)
result = cv2.add(result, add_warped)

# Display the result
cv2.imshow('Superimposed Image', result)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Save the result
cv2.imwrite(r'W:\University of Moratuwa\Academics\Semester 5\EN3160 - Image Processing and Machine Vision\Assignments\Assignment_2\Result.JPG', result)

# Plot the result
result = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
plt.imshow(result)
plt.show()

```

```python
import cv2
import numpy as np

# Load the architectural image and the flag image
architectural_image = cv2.imread(r'W:\University of Moratuwa\Academics\Semester 5\EN3160 - Image Processing and Machine Vision\Assignments\Assignment_2\Bird_Base.JPG')
flag_image = cv2.imread(r'W:\University of Moratuwa\Academics\Semester 5\EN3160 - Image Processing and Machine Vision\Assignments\Assignment_2\Bird_Add.JPG')

# Create a copy of the architectural image for user interaction
architectural_copy = architectural_image.copy()
points = []

# Define a callback function for mouse click events
def select_points(event, x, y, flags, param):
    global points

    if event == cv2.EVENT_LBUTTONDOWN:
        points.append([x, y])
        cv2.circle(architectural_copy, (x, y), 5, (0, 255, 0), -1)
        architectural_copy_r = cv2.resize(architectural_copy, (800, 600))  # Adjust the size as needed
        cv2.imshow('Architectural Image', architectural_copy_r)

        # If four points are selected, compute the homography and superimpose the flag
        if len(points) == 4:
            architectural_points = np.array([[0, 0], [architectural_image.shape[1], 0], [architectural_image.shape[1], architectural_image.shape[0]], [0, architectural_image.shape[0]]], dtype=np.float32)
            flag_points = np.array(points, dtype=np.float32)
            
            # Compute the homography matrix
            homography_matrix, _ = cv2.findHomography(flag_points, architectural_points)

            # Warp the flag image to fit the architectural image
            flag_warped = cv2.warpPerspective(flag_image, homography_matrix, (architectural_image.shape[1], architectural_image.shape[0]))

            # Create a mask for the flag
            flag_mask = np.zeros_like(architectural_image)
            cv2.fillPoly(flag_mask, [architectural_points.astype(int)], (255, 255, 255))

            # Invert the mask
            flag_mask_inv = cv2.bitwise_not(flag_mask)

            # Blend the flag onto the architectural image
            result = cv2.bitwise_and(architectural_image, flag_mask_inv)
            result = cv2.add(result, flag_warped)

            # Resize the result image to fit the screen
            scaled_result = cv2.resize(result, (800, 600))  # Adjust the size as needed

            # Display the scaled result
            cv2.imshow('Superimposed Image', scaled_result)
            

cv2.imshow('Architectural Image', architectural_copy)
cv2.setMouseCallback('Architectural Image', select_points)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
