# Spacial Filtering

Spatial filtering, also known as convolution, is a fundamental concept in image processing and computer vision. It involves applying a mathematical operation to an image by sliding a small filter (also called a kernel or mask) over the image's pixels. The filter is a small matrix that defines how the operation should be performed at each pixel and its neighboring pixels. Spatial filtering is used to perform various image enhancement and transformation tasks, such as blurring, sharpening, edge detection, and noise reduction.

The basic steps of spatial filtering include:

1. **Convolution**: The filter is placed on a pixel in the image, and a mathematical operation (typically multiplication and summation) is performed between the filter values and the corresponding pixel values in the neighborhood. This produces a new value for the central pixel.

2. **Sliding Window**: The filter is then moved to the next pixel, and the process is repeated. This sliding window operation covers the entire image, and the result is a new image with the filter's operation applied to each pixel.

Spatial filters are often used for tasks like:

- **Smoothing/Blurring**: Using filters with averaging or Gaussian distribution to reduce noise and create a smoother version of the image.
- **Sharpening**: Using filters to enhance edges and details in the image.
- **Edge Detection**: Applying filters that highlight regions of rapid intensity change (edges) in the image.
- **Noise Reduction**: Using filters to reduce noise in the image while preserving important features.
- **Embossing**: Applying filters to create a three-dimensional effect or highlight specific image features.
- **Gradient Calculation**: Calculating gradients for edge detection or other purposes.

Different filters are designed for different tasks, and the size and values of the filter matrix determine the specific effect. Common filter sizes include 3x3, 5x5, and 7x7, but other sizes can also be used based on the desired outcome.

Spatial filtering is a key concept in image processing and is the foundation of many advanced techniques used in computer vision and image analysis. It allows you to modify an image's appearance and extract meaningful information by processing its pixel values in a systematic manner.

## Spatial Filtering(Sobel) Using 2D

* Sobel filtering is a commonly used technique in image processing and computer vision for edge detection.

* Edge detection is a fundamental step in many image processing tasks, such as object recognition, image segmentation, and feature extraction.

* The Sobel filter works by convolving an image with a small kernel (usually 3x3 or 5x5) that represents a gradient operator. The gradient of an image indicates the change in intensity or color at each pixel. By calculating the gradient, we can identify regions of rapid intensity change, which typically correspond to edges in the image. 

* There are two main variations of the Sobel filter: one for detecting edges in the horizontal direction and the other for detecting edges in the vertical direction. These filters are usually referred to as the **Sobel-X** and **Sobel-Y** filters, respectively.

* Sobel-X 
[-1  0  1]
[-2  0  2]
[-1  0  1]

* Sobel-Y
[-1 -2 -1]
[0  0  0] 
[1  2  1]

### Filtering Using loops

```python
%matplotlib inline

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import math

def filter( image , kernel ) :

    assert kernel.shape [0]%2 == 1 and kernel . shape [ 1 ]%2 == 1
    
    k_hh , k_hw = math.floor( kernel . shape [ 0 ] / 2 ) , math.floor( kernel . shape [ 1 ] / 2 )
    
    h , w = image . shape
    
    image_float = cv . normalize ( image . astype ( 'float' ) , None , 0.0 , 1.0 , cv.NORM_MINMAX)
    
    result = np . zeros ( image . shape , 'float' )

    for m in range ( k_hh , h - k_hh ) :
        
        for n in range ( k_hw , w - k_hw) :
            result [ i , j ] = np.dot ( image_float [m - k_hh :m + k_hh + 1 , n - k_hw : n + k_hw + 1 ] . flatten( ) , kernel.flatten( ) )
    
    return result

img = cv . imread ( 'C:/Python/ipmv/Scripts/Image_Processing_NoteBooks/Images/C.jpg', cv . IMREAD_REDUCED_GRAYSCALE_8)

f , axarr = plt.subplots ( 1 , 2 )

axarr [ 0 ] . imshow( img , cmap= 'gray')
axarr [ 0 ] . set_title( 'Original')

kernel = np . array ( [ ( 1 / 9 , 1 /9 , 1 /9) , ( 1 / 9 , 1 /9 , 1 /9) , ( 1 / 9 , 1 /9 , 1 /9) ] , dtype= 'float' )

imgb = filter( img , kernel )
imgb = imgb*255.0
imgb = imgb . astype ( np . uint8 )

axarr [ 1 ] . imshow( imgb , cmap= 'gray' )
axarr [ 1 ] . set_title( 'Filtered' )
```

## Sharpening

### Using Gaussian Filter

* Better than the box filter because when using a box filter, the rectangular nature can be seen in the processed image

```python
%matplotlib inline

import cv2 as cv
import numpy as np

original = cv . imread ( 'C:/Python/ipmv/Scripts/Image_Processing_NoteBooks/Images/tom.jpg' , cv.IMREAD_GRAYSCALE)

assert original is not None

smoothened = cv.GaussianBlur(original, (5,5),0).astype(np.float32)

difference = original - smoothened

sharpened = cv.addWeighted(original.astype(np.float32), 1, difference, 1.5,0)

#Display all the images

cv.namedWindow('Images', cv.WINDOW_AUTOSIZE)
cv.imshow('Images', original)
cv.waitKey(0)
cv.imshow('Images', smoothened)
cv.waitKey(0)
cv.imshow('Images', sharpened)
cv.waitKey(0)
cv.imshow('Images', difference)
cv.waitKey(0)
cv.destroyAllWindows()

```
