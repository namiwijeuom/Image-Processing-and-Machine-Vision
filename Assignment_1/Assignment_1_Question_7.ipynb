## Part(b)

* Own code to Sobel filter the image can be implemeted by doing the convolution of the kernel with the image using for loops. 

```python
img = cv . imread ( 'W:/University of Moratuwa/Academics/Semester 5/EN3160 - Image Processing and Machine Vision/Assignments/Assignment_1/Images/einstein.png' , cv.IMREAD_REDUCED_GRAYSCALE_2)

# Define the custom Sobel filters (horizontal and vertical gradients)
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

sobel_y = np.array([[-1, -2, -1],
                    [ 0,  0,  0],
                    [ 1,  2,  1]])


# Convolution operation for gradient computation
def convolve(image, kernel):
    kernel_size = kernel.shape[0]
    print(kernel_size)
    image_height, image_width = image.shape

    print('Image Height : ', image_height)
    print('Image Width : ', image_width)

    # Initialize the result image
    result = np.zeros((image_height - kernel_size + 1, image_width - kernel_size + 1))

    print('Result Image Shape : ', result.shape)
    print('Result : ', result)
    
    # Perform convolution
    for i in range(result.shape[0]):
        for j in range(result.shape[1]):
            result[i, j] = np.sum(image[i:i+kernel_size, j:j+kernel_size] * kernel)

    return result

# Apply the custom Sobel filters using the convolve function
sobel_h = convolve(img, sobel_x)
sobel_v= convolve(img, sobel_y)

# Display the original image and gradient magnitude
plt.figure(figsize=(12, 8))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(np.abs(sobel_h), cmap='gray')
plt.title('Sobel Gradient X')

plt.subplot(1, 3, 3)
plt.imshow(np.abs(sobel_v), cmap='gray')
plt.title('Sobel Gradient Y')

plt.tight_layout()
plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image_path = 'W:/University of Moratuwa/Academics/Semester 5/EN3160 - Image Processing and Machine Vision/Assignments/Assignment_1/Images/einstein.png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define the custom Sobel filters (horizontal and vertical gradients)
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

sobel_y = np.array([[-1, -2, -1],
                    [ 0,  0,  0],
                    [ 1,  2,  1]])

# Convolution operation for gradient computation
def convolve(image, kernel):
    kernel_size = kernel.shape[0]
    
    image_height, image_width = image.shape

    print('Image Height : ', image_height)
    print('Image Width : ', image_width)

    # Initialize the result image
    result = np.zeros((image_height - kernel_size + 1, image_width - kernel_size + 1))
  
    # Perform convolution
    for i in range(result.shape[0]):
        for j in range(result.shape[1]):
            result[i, j] = np.sum(image[i:i+kernel_size, j:j+kernel_size] * kernel)

    return result

# Apply the custom Sobel filters using the convolve function
gradient_x = convolve(image, sobel_x)
gradient_y = convolve(image, sobel_y)

# Compute the gradient magnitude
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Display the original image and gradient magnitude
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(gradient_x, cmap='gray')
plt.title('Sobel Gradient X')

plt.subplot(2, 2, 3)
plt.imshow(gradient_y, cmap='gray')
plt.title('Sobel Gradient Y')

plt.subplot(2, 2, 4)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Gradient Magnitude')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image_path = 'W:/University of Moratuwa/Academics/Semester 5/EN3160 - Image Processing and Machine Vision/Assignments/Assignment_1/Images/einstein.png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define the separable Sobel filters (horizontal and vertical gradients)
sobel_horizontal = np.array([1, 2, 1])
sobel_vertical = np.array([1, 0, -1])

# Convolution operation using separable kernels
def convolve_separable(image, kernel_horizontal, kernel_vertical):
    image_height, image_width = image.shape
    kernel_size = kernel_horizontal.shape[0]

    # Initialize the result image
    result = np.zeros((image_height - kernel_size + 1, image_width - kernel_size + 1))

    # Convolve with horizontal kernel
    for i in range(result.shape[0]):
        for j in range(result.shape[1]):
            result[i, j] = np.sum(image[i:i+kernel_size, j:j+kernel_size] * kernel_horizontal)

    # Convolve with vertical kernel
    final_result = np.zeros_like(result)
    for i in range(result.shape[0]):
        for j in range(result.shape[1]):
            final_result[i, j] = np.sum(result[i:i+kernel_size, j:j+kernel_size] * kernel_vertical)

    return final_result

# Apply the Sobel filter using the convolve_separable function
gradient = convolve_separable(image, sobel_horizontal, sobel_vertical)

# Compute the gradient magnitude
gradient_magnitude = np.abs(gradient)

# Display the original image and gradient magnitude
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Gradient Magnitude')

plt.tight_layout()
plt.show()
```
