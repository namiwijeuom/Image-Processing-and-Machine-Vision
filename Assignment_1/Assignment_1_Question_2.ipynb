# Question 2

Apply a similar operation as above (question 1) to accentuate

1. white matter
2. gray matter in the brain proton density image shown in Fig. 2. Show the intensity transformations as a plots.

```python
#Import necessary libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Linear transformation fuctions for the image enhancement 1
t1 = np.linspace(0, 10, 140 ).astype('uint8')
print(t1)
print(len(t1))

t2 = np.linspace(11, 230 ,90 ).astype('uint8')
print(t2)
print(len(t2))

t3 = np.linspace(231, 255 , 26).astype('uint8')
print(t3)
print(len(t3))

transform = np.concatenate((t1,t2),axis = 0).astype('uint8')
transform_final = np.concatenate((transform,t3),axis = 0).astype('uint8')

#Linear transformation fuctions for the image enhancement 2
ta = np.linspace(0, 1, 120 ).astype('uint8')
print(ta)
print(len(ta))

tb = np.linspace(11, 235 ,25 ).astype('uint8')
print(tb)
print(len(tb))

td = np.linspace(236, 200 , 27).astype('uint8')
print(td)
print(len(td))

tc = np.linspace(199, 0 , 9).astype('uint8')
print(tc)
print(len(tc))

te = np.linspace(0, 1 , 75).astype('uint8')
print(tc)
print(len(tc))

transform_1 = np.concatenate((ta,tb),axis = 0).astype('uint8')
transform_2 = np.concatenate((transform_1,td),axis = 0).astype('uint8')
transform_3 = np.concatenate((transform_2,tc),axis = 0).astype('uint8')
transform_final_2 = np.concatenate((transform_3,te),axis = 0).astype('uint8')

# Read the image
img_orig = cv.imread('W:/University of Moratuwa/Academics/Semester 5/EN3160 - Image Processing and Machine Vision/Assignments/Assignment_1/Images/BrainProtonDensitySlice9.png',cv.IMREAD_GRAYSCALE)

# Apply the transformation - 1
image_transformed = cv.LUT(img_orig,transform_final)

# Apply the transformation - 2
image_transformed_2 = cv.LUT(img_orig,transform_final_2)

# Display the images
plt.figure(figsize=(15, 8))

plt.subplot(1, 3, 1)
plt.imshow(img_orig, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(image_transformed, cmap='gray')
plt.title('Enhanced Image - 1')

plt.subplot(1, 3, 3)
plt.imshow(image_transformed_2, cmap='gray')
plt.title('Enhanced Image - 2')

plt.tight_layout()
plt.show()

# Plot the transformation functions
fig , ax = plt.subplots()

ax.plot(transform_final, color = 'red')
ax.plot(transform_final_2, color = 'blue')
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel(r'Output, $\mathrm{t}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
ax.grid(True)
ax.legend(['Transformation - 1','Transformation - 2'])
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = 'W:/University of Moratuwa/Academics/Semester 5/EN3160 - Image Processing and Machine Vision/Assignments/Assignment_1/Images/BrainProtonDensitySlice9.png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Normalize the image to [0, 1]
normalized_image = image.astype(np.float32) / 255.0

# Linear transformation parameters for white matter and gray matter enhancement
white_matter_scale = 1.2
white_matter_shift = 0.1
gray_matter_scale = 1.0
gray_matter_shift = 0.2

# Apply linear transformations to accentuate white matter and gray matter
white_matter_transformed = np.clip(white_matter_scale * normalized_image + white_matter_shift, 0, 1)
gray_matter_transformed = np.clip(gray_matter_scale * normalized_image + gray_matter_shift, 0, 1)

# Plot the linear transformations
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(normalized_image, white_matter_transformed, 'b', label='White Matter Transformation')
plt.plot(normalized_image, gray_matter_transformed, 'r', label='Gray Matter Transformation')
plt.title('Linear Transformations')
plt.xlabel('Original Normalized Pixel Value')
plt.ylabel('Transformed Normalized Pixel Value')
plt.legend()

plt.subplot(1, 2, 2)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.tight_layout()
plt.show()
```
