# Question 1

Implement the intensity transformation depicted  on the image shown.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

#Linear transformation fuctions

#                 0      1         2         3
#               0  1   0   1     0   1     0   1
c = np.array([(50,50),(50,100),(150,255),(150,150)])
print(c)

t1 = np.linspace(0, c[0,1], c[0,0] ).astype('uint8')
print(len(t1))

t2 = np.linspace(c[1,1] + 1, c[2,1] ,c[2,0] - c[1,0] ).astype('uint8')
print(len(t2))

t3 = np.linspace(c[3,1] + 1, 255 , 255 - c[3,0] + 1 ).astype('uint8')
print(len(t3))

transform = np.concatenate((t1,t2),axis = 0).astype('uint8')
transform_final = np.concatenate((transform,t3),axis = 0).astype('uint8')

fig , ax = plt.subplots()

ax.plot(transform_final)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel(r'Output, $\mathrm{t}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
plt.savefig('transform.png')
plt.show()

img_orig = cv.imread('W:/University of Moratuwa/Academics/Semester 5/EN3160 - Image Processing and Machine Vision/Images/emma.jpg',cv.IMREAD_GRAYSCALE)
print(img_orig.shape)

cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Original',img_orig)
cv.waitKey(0)

image_transformed = cv.LUT(img_orig,transform_final)

cv.imshow('Transformed', image_transformed)
cv.waitKey(0)
cv.destroyAllWindows()

# Display the original images
plt.figure(figsize=(12, 8))

plt.subplot(1, 3, 1)
plt.imshow(img_orig, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(image_transformed, cmap='gray')
plt.title('Transformed Image')

plt.tight_layout()
plt.show()
```
